# Distribution of Accidents Across Different States

```python
state_counts = df['State'].value_counts()

state_counts_df = state_counts.reset_index()
state_counts_df.columns = ['State', 'Number of Accidents']
```

```python
plt.figure(figsize=(12, 8))
plt.bar(state_counts_df['State'], state_counts_df['Number of Accidents'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents Across Different States')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Distribution of Accidents Based on Weather Conditions

```python
weather_counts = df['WeatherCondition'].value_counts()

weather_counts_df = weather_counts.reset_index()
weather_counts_df.columns = ['WeatherCondition', 'Number of Accidents']
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='WeatherCondition', y='Number of Accidents', data=weather_counts_df, palette='viridis')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents Based on Weather Conditions')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

# State-wise Total Death to Accident Ratio

```python
state_totals = df.groupby('State').agg({'Deaths': 'sum', 'State': 'count'}).rename(columns={'State': 'TotalAccidents'})

state_totals['DeathToAccidentRatio'] = state_totals['Deaths'] / state_totals['TotalAccidents']
state_totals = state_totals.reset_index()
print(state_totals)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='State', y='DeathToAccidentRatio', data=state_totals, palette='viridis')
plt.xlabel('State')
plt.ylabel('Death to Accident Ratio')
plt.title('State-wise Total Death to Accident Ratio')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```
