# Handling Missing Values

```python
print(df.isnull().sum())

print(df[df.isnull().any(axis=1)])

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
print(df.duplicated().sum())
```

```python
categorical_columns = ['Education', 'EmploymentType', 'MaritalStatus', 'LoanPurpose']
df[categorical_columns] = df[categorical_columns].astype('category')
```

# Unvariate Analysis

```python
#Distribution of Numerical Features
sns.countplot(x='Default', data=df)
plt.title('Distribution of Loan Default')
plt.show()
```

```python
#Distribution of Numerical Features
df[['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'InterestRate', 'DTIRatio']].hist(figsize=(15, 10), bins=30)
plt.show()
```

```python
#Distribution of Categorical Features
for col in categorical_columns:
    plt.figure(figsize=(10, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

# Bivariate Analysis

```python
# Correlation matrix 
corr_matrix = df[['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'InterestRate', 'DTIRatio']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Categorical Features vs. Target Variable
for col in categorical_columns:
    plt.figure(figsize=(10, 4))
    sns.barplot(x=col, y='Default', data=df)
    plt.title(f'{col} vs. Loan Default')
    plt.show()
```

```python
#Numerical Features vs. Target Variable
numerical_columns = ['Age', 'Income', 'LoanAmount', 'CreditScore', 'MonthsEmployed', 'NumCreditLines', 'InterestRate', 'DTIRatio']
for col in numerical_columns:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x='Default', y=col, data=df)
    plt.title(f'{col} vs. Loan Default')
    plt.show()
```

# Feature Engineering

```python
#Creating a feature for income to loan amount ratio
df['IncomeToLoanRatio'] = df['Income'] / df['LoanAmount']
```

```python
# Convert categorical variables to dummy variables
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```
